```python
def and_rule(a,b,c,d,e,x, y, z):
    rule = np.fmax(a, b)
    rule = np.fmax(rule, c)
    rule = np.fmax(rule, d)
    rule = np.fmax(rule, e)
    rule = np.fmax(rule, x)
    rule = np.fmax(rule, y)
    act = np.fmax(rule, z)
    return act

def or_rule(x, y, z):
    rule = np.fmax(x, y)
    act = np.fmax(rule, z)
    return act
```
